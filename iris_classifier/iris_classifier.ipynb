```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
%matplotlib notebook
```

```python
dataset=pd.read_csv('iris.csv',header=None)
```

```python
#Any missing values
dataset.isnull().sum()
```

```python
#size of dataset
dataset.shape
dataset[4].unique()
```

```python
is_set=dataset[dataset[4]=='Iris-setosa']
iv_set=dataset[dataset[4]=='Iris-versicolor']
ivi_set=dataset[dataset[4]=='Iris-virginica']
```

```python
#plotting graphs
plt.scatter(is_set[0],is_set[1],c='red',label='Iris_setosa')
plt.scatter(iv_set[0],iv_set[1],c='green',label='Iris-versicolor')
plt.scatter(ivi_set[0],ivi_set[1],c='blue',label='Iris-virginica')
plt.xlabel('sapel-length')
plt.ylabel('sapel-width')
plt.legend()
```

```python
#petal-length vs petal-width
plt.scatter(is_set[2],is_set[3],c='red',label='Iris_setosa')
plt.scatter(iv_set[2],iv_set[3],c='green',label='Iris-versicolor')
plt.scatter(ivi_set[2],ivi_set[3],c='blue',label='Iris-virginica')
plt.xlabel('sapel-length')
plt.ylabel('sapel-width')
plt.legend()
```

```python
#Independent and dependent variables
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#standardizing the dataset
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
#applying classification algorithms
from sklearn import metrics #to find accuracy

#logistic Regression
from sklearn.linear_model import LogisticRegression
lr_classifier=LogisticRegression()
lr_classifier.fit(x_train,y_train)
y_pred=lr_classifier.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

```python
#svm
from sklearn.svm import SVC
svm_classifier=SVC()
svm_classifier.fit(x_train,y_train)
svm_y=lr_classifier.predict(x_test)
metrics.accuracy_score(svm_y,y_test)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dtc_classifier=DecisionTreeClassifier()
dtc_classifier.fit(x_train,y_train)
dtc_y=dtc_classifier.predict(x_test)
metrics.accuracy_score(dtc_y,y_test)
```

```python
#kNN
from sklearn.neighbors import KNeighborsClassifier
kn_classifier=KNeighborsClassifier(n_neighbors=5)
kn_classifier.fit(x_train,y_train)
kn_y=kn_classifier.predict(x_test)
metrics.accuracy_score(kn_y,y_test)
```

```python
#random forest classification
from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
rf_classifier.fit(x_train,y_train)
rf_y=rf_classifier.predict(x_test)
metrics.accuracy_score(y_test,rf_y)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
nb_classifier=GaussianNB()
nb_classifier.fit(x_train,y_train)
nb_y=nb_classifier.predict(x_test)
metrics.accuracy_score(y_test,nb_y)
```
